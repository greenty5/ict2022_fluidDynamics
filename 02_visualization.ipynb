First of all, we have to import python laibraries which are really important in this course.

````python
import numpy as np # fundamental package for scientific computing
import matplotlib.pyplot as plt # visualization package
````

Here, let us try to make a graph taht indicates sine curve. 

Firstly, let us make two variables for the sine curve.

````python
x = np.linspace(0, 2*3.1415, 100)
y = np.sin(x)
````

Now we can make a graph using plot function as shown in below:

````python
plt.plot(x,y)
````

Next, let us make scatter plot.

````python
np.random.seed(2022)
x2 = np.random.uniform(0,100,20)
y2 = x2*np.random.uniform(1,2,20)
````

````python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_subplot(2,1,1)
ax1.plot(x,y)
ax1.set_xlabel('time [s]')
ax1.set_ylabel('u [m/s]')

ax2 = fig.add_subplot(2,1,2)
ax2.scatter(x2, y2)
ax2.set_xlabel("weight [g]")
ax2.set_ylabel('lenght [cm]')
ax2.set_xlim(0,110)
ax2.set_ylim(0,190)

fig.tight_layout()
fig.show()
````

where, 
```python
ax = fig.add_subplot(2,1,i)
```
_2,1_ denote the number of graph we want to draw simultaneously, and _i_ denotes the oder of the graph. For another example is ...

````python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y)
ax1.set_xlabel('time [s]')
ax1.set_ylabel('u [m/s]')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x2, y2)
ax2.set_xlabel("weight [g]")
ax2.set_ylabel('lenght [cm]')
ax2.set_xlim(0,110)
ax2.set_ylim(0,190)

fig.tight_layout()
fig.show()
````
