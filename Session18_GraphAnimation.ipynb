### Session 1.8 How to make moving graph (like animation) by using matplotlib

In some cases, it should be better to express our fluid dynamcis calculatio nresults using animation. 
We will learn learn how make moving graph (like animation) by matplotlib. 

Please note that this course does not recommend the animation technique too much, even though the animation is a powerful and excellent material to express our calculation results. Because making the animations is too heavy on the Google colaboratory.

---

Jupyter notebook can display animation by importing `matplotlib.animation` and then `%matplotlib nbagg` for interactive image display on Jupyter.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

```

Since the `nbagg` function is not available on Google Colab, it is necessary to add code to run `javascript` instead. The code mentioned below is for display an animate with the random number generation based on `ArtistAnimation`.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

x = np.linspace(0,2*np.pi, 100) 

# making fig, ax objects
fig, ax = plt.subplots()

# making graph lists
# brabrabra

# labels of x and y axes
ax.set_xlabel(r"$x$", fontsize=15)
ax.set_ylabel(r"$y$", fontsize=15)
# setting ranges of x and y axes
ax.set_xlim([0, 2*np.pi])
ax.set_ylim([-1.5, 1.5]) 

# Substituting `fig` object and `ims` list into ArtistAnimation, and making animation
# bra bra


# Google Colaboratory
rc('animation', html='jshtml')
plt.close()
anim
````

`from IPython.display import HTML` is needed to display html on Jupyter, `rc` is a function to format the plot: [matplotlib.pyplot.rc - Matplotlib 3.5.1](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.rc.html). You can run javascript by setting html to jshtml in the documentation.

When this is executed, a graph with a slider is displayed. And when the play button is clicked, an animation is displayed.

````python
size = 100
x = np.arange(size)
fig, ax = plt.subplots()
plt.close()
ax.set_xlim(( 0, size))
ax.set_ylim((-3, 3))
line, = ax.plot([],[],'b') # 'b' means blue.

def animate(i):
  yrand=np.random.randn(size)  
  line.set_data(x,yrand)
  return (line)

anim = animation.FuncAnimation(fig, animate, frames=10, interval=100)
rc('animation', html='jshtml')
anim
````
