In this session, let us learn how to draw contour graph by matplotlib. 

__contour graph is very useful to descrive fluid flow fied, especially pressure distribution and velocity magnitude__.

First of all, we have to import python laibraries as we learned the previous session.

````python
import numpy as np # fundamental package for scientific computing
import matplotlib.pyplot as plt # visualization package
````

The function for contour graph drawing is __matplotlib.pyplot.[contour](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html)__. The function draw contour lines and filled contours. 
Call signature is ...

```python
matplotlib.pyplot.contour([X, Y], Z, [levels], **kwargs)
```

where X, Y define x and y coordinates, must both be 2D with the same shape as Z, and Z is the height values.

The argument __levels__ is to determine the number and positions of the contour lines / regions.

Here we will make sample function like ...

````python
def f(x, y):
  return x**2 + y**2 + x*y
````

Making coordinate matrix for _f(x,y)_ function; 

````python
x = np.arange(-3,4)
y = np.arange(-3,4)
X, Y = np.meshgrid(x,y)
````

Calculating height data (__Z__);

````python
Z = f(X,Y)
````

Now, graph data is complete. Let us draw a contour graph;

````python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.set_title("contour")
ax.grid()
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.contour(X,Y,Z)
plt.show()
````

You may find that this graph is coarse ...

The reason why the coarse graph is the number of graph data. In the above graph, we prepared only 3 x 3 like;

here `cmap = 'PuOr` means that __Pu__ (purple) indicates positive value, and __Or__ (orange) indicates negative value.

And also we can add labels on the lines in the contour graph.

````python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
cont10 = ax.contour(X,Y,Z, 10, cmap='PuOr', linestyles=["-", "--", "-", "--", "-"])
ax.set_title("contour")
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.clabel(cont10, inline=True, fontsize=8)
plt.colorbar(cont10)

plt.show()
````

if we want to filled contours, we can use __[contourf](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html)__ function.

````python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
cont10 = ax.contourf(X,Y,Z, 10, cmap='Blues', linestyles=["-", "--", "-", "--", "-"])
ax.set_title("contour")
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
#ax.clabel(cont10, inline=True, fontsize=8)
plt.colorbar(cont10)

plt.show()
````
